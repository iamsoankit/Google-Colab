#Arithemetic Operator : +,-,/,*,%,//,**
#Assignment Operator : +=,-=,*=,/=,%=,//=,**=
#Relational/Comparison Operators : ==,!=,<,>,<=,>=
#Logical Operators : and , not ,or

```python
#Example of Arithemetic operator

a = 20
b = 10


print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b) #(Remainder of the division)
print(a//b) #(Quotient of the division)
print(a**b) #(a raised to b)
```

```python
#WAP a program to take average of three numbers

a1=5
a2=6
a3=8

average = ((a1+a2+a3)/3)

print(average)
```

```python
#Example of Assignment operator

a1=7
a1=a1+6 #or a1 +=6

print(a1)
```

```python
#Example of Relational operator

a=80
b=96

print('a > b', a>b)
print('a < b', a<b)
print('a >= b',a>=b)
print('a <= b', a<=b)
print('a == b',a==b)
print('a != b',a!= b) #not equal
```

```python
#Example of Logical Operators

x = True
y = False

print('x and y is',x and y)
print('x or y is',x or y)
print('not x is',not x)

```
